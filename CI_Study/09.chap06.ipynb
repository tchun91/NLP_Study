Question 1: Weighted Regression Application
Question: When analyzing data that has been grouped due to privacy concerns, why might a researcher choose to use a weighted least squares regression instead of ordinary least squares?

It means it shows non constant variance  across all values of the features. (ex) higher age higher income). Grouped data can be very messed up.

Question 3: Role of Dummy Variables
Question: Why are dummy variables used in regression models, and what is a potential pitfall if the base category is not properly chosen?

Dummy variables are used to calculate treated and untreated means. It's working like a switch. 

Question 4: Interpreting Coefficients with Dummy Variables
Question: How do you interpret the coefficient of a dummy variable in a regression model that also includes continuous variables?

Question 5: Comparing Parametric and Nonparametric Regression Models
Question: What are the advantages and disadvantages of using a nonparametric model (like dummy variables for each category) compared to a parametric model in regression analysis?

If an effect is continuous meaning parametric, the effect will be a single parameter meaning we can only check the overall effect. if we change it to dummy variable, it becomes partition and we can check each coef of the group. However, we will lose statistical significance.


```python

import pandas as pd
import numpy as np

np.random.seed(200)

n = 1000
data = {
    'Marketing_Spend': np.random.gamma(shape=2, scale=50, size=n),
    'Store_Size': np.random.uniform(1, 20, n),
    'Staff_Numbers': np.random.poisson(lam=30, size=n),
    'Competition': np.random.randint(1, 11, n),
    'Economic_Condition': np.random.randint(1, 11, n)
}

df = pd.DataFrame(data)

df['Monthly_Sales'] = (df['Marketing_Spend'] * 1.5 +
                       df['Store_Size'] * 10 +
                       df['Staff_Numbers'] * 2 -
                       df['Competition'] * 3 +
                       df['Economic_Condition'] * 5 +
                       np.random.normal(0, 50, n))

print(df.head())
```

```python
#Question 1: Fit a multiple regression model to predict Monthly_Sales using all available independent variables. Identify which independent variable is the most significant predictor of sales and explain its impact.
import warnings
warnings.filterwarnings('ignore')

import pandas as pd
import numpy as np
from scipy import stats
from matplotlib import style
import seaborn as sns
from matplotlib import pyplot as plt
import statsmodels.formula.api as smf

style.use("fivethirtyeight")

m1 = smf.wls('Monthly_Sales ~ Marketing_Spend + Store_Size + Staff_Numbers + C(Competition) + C(Economic_Condition) + Marketing_Spend',data=df,weights = df['Staff_Numbers']).fit()
m1.summary().tables[1]

# Economic condition when T = 7 has the biggest impact. Also Economic_condition in the data context has the second highest impact to the prediction.

```

```python
#Question 2: Add interaction terms between Marketing_Spend and each of the variables Economic_Condition and Competition. What do these interactions indicate about the effects of marketing efforts under different economic and competitive conditions?
m2 = smf.wls('Monthly_Sales ~ Marketing_Spend*C(Economic_Condition) + Marketing_Spend*C(Competition) + Store_Size + Marketing_Spend',data=df,weights = df['Staff_Numbers']).fit()
m2.summary().tables[1]

#Interaction makes things more narrower and details. It gives the conditional treatment effect. For example when EC = 9, when marketing increase by 1 unit, the monthly sales increases by -0.2436.
```
